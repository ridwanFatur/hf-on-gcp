```python
import os
from google.cloud import aiplatform
from dotenv import load_dotenv
```

```python
load_dotenv() 
```

```python
PROJECT_ID = os.environ["PROJECT_ID"]
LOCATION = os.environ["LOCATION"]
```

```python
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = r"./service_account.json"
```

```python
aiplatform.init(
    project=PROJECT_ID,
    location=LOCATION,
)
```

```python
models = aiplatform.Model.list()
```

```python
if not models:
    print("No models found in the Model Registry.")
else:
    print(f"Found {len(models)} models:")
    for model in models:
        print(f"Name: {model.display_name}")
        print(f"ID: {model.resource_name}")
        print(f"Version: {model.version_id}")
        print("-" * 30)
```
